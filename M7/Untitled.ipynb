```python
## financial , 
# consulting,
# tech,
# cpg,
# health,
# realestate,
# mnufacturing,
# media and enter,
# energy,
# gov

Post = {
    "UOP":[37,25,15,6,6,4,2,2,0,0],
    "GSB":[31,18,33,2,5,1,0,3,1,0],
    "HBS":[29,25,19,6,6,3,5,4,0,0],
    "MITS":[16,32,27,3,0,0,19,2,0,2],
    "UOC":[32,31,20,5,0,2,1,2,2,0],
    "CU":[32,34,12,0,2,4,7,4,0,2],
    "NUK":[14,30,28,10,7,3,4,1,1,1],
    "UOC-B":[14,24,32,8,5,2,4,2,2,4],
    "YU":[23,35,15,9,4,2,2,2,2,3],
    "DF":[16,32,28,8,9,0,2,0,1,2],
    "MR":[16,30,0,8,7,0,8,0,0,0],
    "DT":[20,30,24,6,8,0,5,2,0,0],
    "NYU":[34,28,17,7,3,2,0,2,1,1],
    "VD":[26,32,18,8,3,2,4,0,2,0],
    "CJ":[29,22,18,10,6,2,6,0,1,0],
    "UCLA":[18,14,32,11,7,4,0,8,2,2],
    "CMT":[11,23,32,10,12,0,6,2,1,0],
    "USC":[16,30,19,8,6,3,1,9,3,0],
    "UKF":[20,15,21,8,9,11,2,0,6,0],
    "TM":[18,21,28,9,3,5,3,2,5,3]
}


def getShortedByGMAT(GMAT):
    temp = {}
    li = []
    for i in range(len(dataset)):
        temp[dataset[i][1] - GMAT] = dataset[i][0]
        
    for i in sorted(temp):
        li.append(temp[i])
        
    return li

def getShortedByREGION(REGION):
    
    ind = {
        "E":0,
        "W":1,
        "N":2,
        "S":3,
        "NE":4,
        "NW":5,
        "WS":6,
        "SE":7
    }
    
    dirrr = {
        0 : "E",
        1 : "W",
        2 : "N",
        3 : "S",
        4 : "NE",
        5 : "NW",
        6 : "WS",
        7 : "SE"
    }
    
    index = ind[REGION]
    li = [0]*8
    for i in range(len(direc[index])):
        li[i] += direc[index][i]
        
    dic = {}
    for i in range(8):
        val = li[i]
        dirr = dirrr[i]
        direction_have = []

        for j in range(len(dataset)):
            if dataset[j][2] == dirr:
                direction_have.append(dataset[j][0])
        
        tempp = []
    
        if li[i] in dic:
            tempp = dic[li[i]]
            
        tempp.extend(direction_have)
            
        dic[li[i]] = tempp
    
    returninglist = []
    for i in sorted(dic):
        for j in range(len(dic[i])):
            returninglist.append(dic[i][j])
    return returninglist


def getShortedByPOSTMBA(POST_MBA):
    
    dictt = {}
    
    for i in Post:
        temp = []
        tempp = Post[i][POST_MBA]
        if tempp in dictt:
            temp = dictt[Post[i][POST_MBA]]
        
        temp.append(i)
        dictt[Post[i][POST_MBA]] = temp
        
    returninglist = []
    for i in sorted(dictt,reverse = True):
        for j in range(len(dictt[i])):
            returninglist.append(dictt[i][j])
        
    return returninglist
```
